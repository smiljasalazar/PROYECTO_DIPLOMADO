## 2Ô∏è‚É£ Cargar Modelo Zero-Shot

```python
from transformers import pipeline

# Modelo multiling√ºe para clasificaci√≥n zero-shot
clasificador = pipeline(
    "zero-shot-classification", 
    model="MoritzLaurer/mDeBERTa-v3-base-mnli-xnli"
)

print("‚úÖ Modelo cargado")
```

## 3Ô∏è‚É£ Definir Etiquetas de Triaje

```python
# Etiquetas en espa√±ol
etiquetas_triaje = [
    "Urgencia Vital (Roja)",
    "Consulta Administrativa",
    "Renovaci√≥n de Receta",
    "Consulta M√©dica No Urgente"
]

# Etiquetas en ingl√©s
triage_labels = [
    "Life-Threatening Emergency (Red)",
    "Administrative Inquiry",
    "Prescription Renewal",
    "Non-Urgent Medical Consultation"
]

print("‚úÖ Etiquetas definidas")
```

## 4Ô∏è‚É£ Mensajes de Prueba

```python
mensajes_pacientes = [
    "Tengo un dolor muy fuerte en el pecho y se me duerme el brazo izquierdo.",
    "Hola, necesito saber si el doctor Mart√≠nez atiende los jueves.",
    "Se me acab√≥ la pastilla de la presi√≥n, necesito la receta para comprarla.",
    "Mi hijo tiene fiebre de 38 desde ayer y llora mucho, pero come bien.",
    "Quiero cancelar mi cita de ma√±ana, no podr√© asistir.",
    "Siento una presi√≥n muy fuerte en el pecho y me falta el aire",
]

print(f"‚úÖ {len(mensajes_pacientes)} mensajes de prueba")
```

## 5Ô∏è‚É£ Clasificar Mensajes

```python
print("üîç Clasificando mensajes...\n")
print("="*60)

for mensaje in mensajes_pacientes:
    resultado = clasificador(mensaje, etiquetas_triaje)
    
    print(f"üì© Mensaje: {mensaje[:50]}...")
    print(f"üè∑Ô∏è  Categor√≠a: {resultado['labels'][0]}")
    print(f"üìä Confianza: {resultado['scores'][0]:.1%}")
    print("-"*60)
```

## 6Ô∏è‚É£ Prueba Interactiva

```python
# Escribe tu propio mensaje
mi_mensaje = "Me duele mucho la cabeza y veo luces"  # üëà Cambia esto

resultado = clasificador(mi_mensaje, etiquetas_triaje)

print(f"üì© Tu mensaje: {mi_mensaje}")
print(f"\nüè∑Ô∏è Clasificaci√≥n: {resultado['labels'][0]}")
print(f"üìä Confianza: {resultado['scores'][0]:.1%}")

# Mostrar todas las probabilidades
print("\nüìã Todas las probabilidades:")
for label, score in zip(resultado['labels'], resultado['scores']):
    bar = "‚ñà" * int(score * 20)
    print(f"  {label}: {score:.1%} {bar}")
```
